```python
#in this code cell im taking the original file that is a csv and converting it into a sort of file that I can manipulate it in python 
sales=pd.read_csv('../input/bakery-sales/Bakery Sales.csv')
prices=pd.read_csv('../input/bakery-sales/Bakery price.csv')
```

```python
#the .shape function tells us the shape of the file. For example the sales dataframe has 27 colums and 2654 rows
print(sales.shape)
print(prices.shape)
```

```python
#the tail funtion shows us the last few rows of data and here is when we run into our first issue
#there is a lot of missing data and we have to figure out what to do with the missing values
sales.tail()
```

```python
#We saw in the previous output that there are many rows that simply include no data at all.
#These rows should be dropped since they have no useful information. Not even a date
#dropna is a built in method to delete all rows that have a certain amount of "NaN" values
#I choose to delete rows where all values are "NaN" by setting how="all"
sales=sales.dropna(how='all')
```

```python
#By simply dropping "NaN" values we have reduced the size of the data by just over 200 rows 
sales.shape
```

```python
#We can see here that many rows still have numerous values with"NaN" values
#We can see from the date that in all of these cases "NaN" is used to represent cases in which the value should say 0
sales.tail()
```

```python
#with this function i'm able to replace all values that currently say "NaN" to values that are 0
sales=sales.fillna(value=0)
```

```python
#mission success!!!! We have replaced all "Nan" values with 0
sales.tail()
```

```python
#We find something interesting here on 2020-06-18 for some reason the value under day of week is 0
#We should correct that and see if any other rows have that same issue
sales.tail()
```

```python
#here we create a new dataframe called by_zero that consisted of all rows where the value under day of week is 0
by_zero=sales.loc[sales['day of week'] == 0]
```

```python
#luckily the data only consists of one row that is filled with all zeroes
print(by_zero)
```

```python
#here we are going to replace the value 0 with Thru
sales["day of week"]=sales["day of week"].replace(0, "Thru")
```

```python
#The change looks good!!!!
sales.tail()
```

```python
#we saw in the previous output that each row represents a single transaction 
#we are instead going to group the rows by the date
#we are going to do this by seperate datetime into 2 diffrent columns
#one consisting of date and the other consisting of time
sales['time'] = sales['datetime'].str[10:16]
sales['date'] = sales['datetime'].str[0:10]
sales.tail()
```

```python
#here we are creating a new dataframe that will consist of only diffrent kinds of baked goods, the date, and the day 
#the reason for this is so we can group rows together in order to see how the business performs per day
#and if there is any cyclicality to the business
sales_date=sales.drop(columns=['time','datetime', 'place'])
```

```python
#in this codecell we will be grouping rows together by the date and taking the sum of all the baked goods on that day
sales_date_groupby=sales_date.groupby(['date']).sum()
sales_date_groupby.head()
```

```python
#we have managed to drastically reduce the size of the data and made it much more usable for analysis
#we can continue in python but at this point i will move to a spreadsheet software 
sales_date_groupby.shape
```

```python
#this is the code to download the dataframe as a csv file
sales_date_groupby.to_csv('sales_date_groupby')
```
